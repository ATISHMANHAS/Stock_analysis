```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB, BernoulliNB
from sklearn.metrics import accuracy_score
from imblearn.over_sampling import SMOTE

# 1. Load the mobile price.csv dataset into a DataFrame
df = pd.read_csv('mobile price.csv')

# 2. Check the data type of each column and change it according to their data
print("Data types before conversion:")
print(df.dtypes)

# Example: Convert specific columns to appropriate types
# df['column_name'] = df['column_name'].astype('desired_type')

# 3. Find missing values and treat them if you found any
print("Missing values in each column:")
print(df.isnull().sum())

# Example: Fill missing values with the mean (for numeric columns)
# df.fillna(df.mean(), inplace=True)  # Adjust based on your dataset

# 4. Plot bar graph for target variable and their categories with help of seaborn
target_variable = 'price_range'  # Replace with the actual target variable name
plt.figure(figsize=(10, 6))
sns.countplot(x=target_variable, data=df)
plt.title('Distribution of Target Variable')
plt.xlabel(target_variable)
plt.ylabel('Count')
plt.show()

# Check for class imbalance
target_counts = df[target_variable].value_counts()
print("Target variable distribution:")
print(target_counts)

# If the data is not balanced, treat them using SMOTE
if target_counts.min() < target_counts.max():
    print("Data is imbalanced. Applying SMOTE...")
    smote = SMOTE()
    X = df.drop(target_variable, axis=1)  # Independent variables
    y = df[target_variable]  # Dependent variable
    X_resampled, y_resampled = smote.fit_resample(X, y)
    df_resampled = pd.DataFrame(X_resampled, columns=X.columns)
    df_resampled[target_variable] = y_resampled
else:
    df_resampled = df  # If no imbalance, continue with original data

# 5. Extract independent variables (Xs) and dependent variables (Ys) into separate data objects
X = df_resampled.drop(target_variable, axis=1)
y = df_resampled[target_variable]

# 6. Split data into train and test DataFrame and set test size as 0.27
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.27, random_state=42)

# 7. Build Gaussian and Bernoulli models, compare their accuracy
# Gaussian Naive Bayes
gnb = GaussianNB()

gnb.fit(X_train, y_train)
y_pred_gnb = gnb.predict(X_test)
accuracy_gnb = accuracy_score(y_test, y_pred_gnb)
print(f'Gaussian Naive Bayes Accuracy: {accuracy_gnb:.2f}')

# Bernoulli Naive Bayes (ensure that features are binary)
# You may need to preprocess X for BernoulliNB
bnb = BernoulliNB()
bnb.fit(X_train, y_train)
y_pred_bnb = bnb.predict(X_test)
accuracy_bnb = accuracy_score(y_test, y_pred_bnb)
print(f'Bernoulli Naive Bayes Accuracy: {accuracy_bnb:.2f}')
```
