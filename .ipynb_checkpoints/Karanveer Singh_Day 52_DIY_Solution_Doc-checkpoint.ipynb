```python
# Import necessary libraries
import pandas as pd
from sklearn.preprocessing import StandardScaler
from scipy.cluster.hierarchy import dendrogram, linkage, cut_tree
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt

# Step 1: Load the dataset
data = pd.read_csv("Country-data.csv")

# Step 2: Create a DataFrame with only numeric data by removing the "country" column
numeric_data = data.drop(columns=["country"])

# Step 3: Scale the data using Standard Scaler
scaler = StandardScaler()
scaled_data = scaler.fit_transform(numeric_data)
scaled_df = pd.DataFrame(scaled_data, columns=numeric_data.columns)

# Step 4: Plotting dendrograms with the complete linkage method
linkage_matrix = linkage(scaled_df, method='complete')
plt.figure(figsize=(10, 7))
dendrogram(linkage_matrix, labels=data["country"].values, leaf_rotation=90)
plt.title("Dendrogram using Complete Linkage")
plt.xlabel("Country")
plt.ylabel("Distance")
plt.show()

# Step 5: Creating cluster labels using cut tree
# Cutting the tree at a specific number of clusters
clusters = cut_tree(linkage_matrix, n_clusters=4).flatten()  # Adjust n_clusters as needed
scaled_df['cluster'] = clusters

# Step 6: Perform the 4-Component PCA on DataFrame
pca = PCA(n_components=4)
pca_result = pca.fit_transform(scaled_df.drop(columns=['cluster']))
pca_df = pd.DataFrame(pca_result, columns=[f'PC{i+1}' for i in range(4)])
pca_df['cluster'] = clusters

# Step 7: Analyze the relationship between GDP and child mortality rate
# Adding back the GDP and child mortality columns for analysis
scaled_df['GDP'] = numeric_data['GDP']  # Ensure the GDP column exists
scaled_df['child_mortality'] = numeric_data['child_mortality']  # Ensure the column exists

# Analyzing low GDP vs. child mortality
low_gdp_df = scaled_df[scaled_df['GDP'] < scaled_df['GDP'].median()]  # Select countries with lower GDP
plt.scatter(low_gdp_df['GDP'], low_gdp_df['child_mortality'])
plt.title("Child Mortality vs. GDP (Low GDP Countries)")
plt.xlabel("GDP")
plt.ylabel("Child Mortality Rate")
plt.show()
```
