```python
import pandas as pd

file_path = 'Groceries_TS_And_ARM_dataset.csv'
df = pd.read_csv(file_path)

df
```

```python
df['Date'] = pd.to_datetime(df['Date'])

df['Year'] = df['Date'].dt.year
df['Month'] = df['Date'].dt.month
df['Day'] = df['Date'].dt.day
df['DayOfWeek'] = df['Date'].dt.day_name()  # Extract day of the week

print(df[['Date', 'Year', 'Month', 'Day', 'DayOfWeek']].head())
```

```python
sales_per_day_of_week = df.groupby('DayOfWeek').size().sort_index()

import matplotlib.pyplot as plt

sales_per_day_of_week.plot(kind='bar', color='blue', figsize=(8, 5))
plt.title('Sales on Each Day of the Week')
plt.xlabel('Day of the Week')
plt.ylabel('Number of Sales')
plt.xticks(rotation=45)
plt.show()
```

```python
sales_per_month = df.groupby('Month').size()

sales_per_month.plot(kind='bar', color='green', figsize=(8, 5))
plt.title('Sales in Each Month of the Year')
plt.xlabel('Month')
plt.ylabel('Number of Sales')
plt.xticks(rotation=0)
plt.show()
```

```python
# Group by Day to observe sales
sales_per_day_of_month = df.groupby('Day').size()

# Plot the results
sales_per_day_of_month.plot(kind='line', marker='o', color='orange', figsize=(10, 5))
plt.title('Sales on Each Day of the Month')
plt.xlabel('Day of the Month')
plt.ylabel('Number of Sales')
plt.grid()
plt.show()
```

```python
if 'Item' in df.columns and 'VisitorID' in df.columns:
    aggregated_data = df.groupby('Date').agg({
        'Item': lambda x: list(set(x)),  
        'VisitorID': 'nunique'          
    }).reset_index()
    aggregated_data.rename(columns={'Item': 'Unique_Items', 'VisitorID': 'Unique_Visitors'}, inplace=True)
    
    print(aggregated_data.head())
else:
    print("The required columns ('Item' and 'VisitorID') are not in the dataset.")
    aggregated_data = df.groupby('Date').size().reset_index(name='Sales')
    print("Aggregated data using available columns:")
    print(aggregated_data.head())
```

```python
if 'Date' in aggregated_data.columns and 'Unique_Visitors' in aggregated_data.columns:
    # Plot trend of visitors over time
    plt.plot(aggregated_data['Date'], aggregated_data['Unique_Visitors'], color='purple')
    plt.title('Number of Visitors Over Time')
    plt.xlabel('Date')
    plt.ylabel('Number of Visitors')
    plt.grid()
    plt.show()
else:
    print("'Date' or 'Unique_Visitors' column is missing in the aggregated data.")
```

```python
from apyori import apriori

if 'TransactionID' in df.columns and 'Item' in df.columns:
    # Prepare transactions data
    transactions = df.groupby('TransactionID')['Item'].apply(list).tolist()

    # Apply Apriori algorithm
    rules = apriori(transactions, min_support=0.0003, min_confidence=0.01, min_lift=3)
    rules_list = list(rules)

    # Sort rules by lift
    sorted_rules = sorted(rules_list, key=lambda x: x.lift, reverse=True)

    # Display top rules
    for rule in sorted_rules[:10]:  # Show top 10 rules
        print(f"Rule: {rule.items}")
        print(f"Support: {rule.support}")
        print(f"Confidence: {rule.ordered_statistics[0].confidence}")
        print(f"Lift: {rule.lift}")
        print("-" * 20)
else:
    print("Required columns ('TransactionID', 'Item') are missing.")
```
