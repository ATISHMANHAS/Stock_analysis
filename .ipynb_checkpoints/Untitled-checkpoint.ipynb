```python
import numpy as np
import scipy.stats as stats

x = int(input("Enter the number of highly satisfied customers: "))
n = 200 
p_hat = x / n
z_critical = stats.norm.ppf(0.95)
se = np.sqrt((p_hat * (1 - p_hat)) / n)
margin_of_error = z_critical * se
lower_bound = p_hat - margin_of_error
upper_bound = p_hat + margin_of_error
print(f"{lower_bound:.5f} {upper_bound:.5f}")
```

```python
import scipy.stats as stats

X = float(input("Enter the speed X (km/hr): "))
mean_speed = 75  
std_dev_speed = 15  
probability = 1 - stats.norm.cdf(X, mean_speed, std_dev_speed)
print(f"{probability:.4f}")
```

```python
import pandas as pd
from sklearn.datasets import load_wine

wine = load_wine()
df = pd.DataFrame(data=wine.data, columns=wine.feature_names)
sample = df.sample(n=50, random_state=100)

mean_alcohol = sample['alcohol'].mean()
std_alcohol = sample['alcohol'].std()
z_critical = stats.norm.ppf(0.975)  
margin_of_error = z_critical * (std_alcohol / np.sqrt(50))
lower_bound = mean_alcohol - margin_of_error
upper_bound = mean_alcohol + margin_of_error
print(f"Z-critical: {z_critical}")
print(f"Margin of Error: {margin_of_error}")
print(f"Confidence Interval: ({lower_bound}, {upper_bound})")
```

```python
import pandas as pd
from sklearn.datasets import load_wine

wine = load_wine()
df = pd.DataFrame(data=wine.data, columns=wine.feature_names)
sample = df.sample(n=50, random_state=100)

mean_alcohol = sample['alcohol'].mean()
std_alcohol = sample['alcohol'].std()
z_critical = stats.norm.ppf(0.975)  
margin_of_error = z_critical * (std_alcohol / np.sqrt(50))
lower_bound = mean_alcohol - margin_of_error
upper_bound = mean_alcohol + margin_of_error
print(f"Z-critical: {z_critical}")
print(f"Margin of Error: {margin_of_error}")
print(f"Confidence Interval: ({lower_bound}, {upper_bound})")
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import binom

n = 10  # number of trials
p = 0.5  # probability of success
k = np.arange(0, n+1)

# Binomial distribution
probabilities = binom.pmf(k, n, p)

plt.bar(k, probabilities)
plt.title('Binomial Distribution (n=10, p=0.5)')
plt.xlabel('Number of successes')
plt.ylabel('Probability')
plt.show()
```

```python
from scipy.stats import poisson

lambda_ = 3  # average rate of occurrence
k = np.arange(0, 11)

# Poisson distribution
probabilities = poisson.pmf(k, lambda_)

plt.bar(k, probabilities)
plt.title('Poisson Distribution (=3)')
plt.xlabel('Number of events')
plt.ylabel('Probability')
plt.show()
```

```python
from scipy.stats import norm

mu = 0  # mean
sigma = 1  # standard deviation
x = np.linspace(-4, 4, 100)

# Normal distribution
probabilities = norm.pdf(x, mu, sigma)

plt.plot(x, probabilities)
plt.title('Normal Distribution (=0, =1)')
plt.xlabel('Value')
plt.ylabel('Probability Density')
plt.show()
```

```python
lambda_ = 1  # rate parameter
x = np.linspace(0, 5, 100)

# Exponential distribution
probabilities = lambda_ * np.exp(-lambda_ * x)

plt.plot(x, probabilities)
plt.title('Exponential Distribution (=1)')
plt.xlabel('Value')
plt.ylabel('Probability Density')
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd

# Data for primary growth (total weight lifted)
data = {
    'Week': [1, 2, 3, 4, 5, 6],
    "Total_Weight": [50, 100, 150, 200, 250, 300],
    "Estimated_1RM": [60, 120, 180, 240, 300, 360]  # Example secondary growth data (Estimated 1RM)
}

# Convert data to DataFrame
df = pd.DataFrame(data)

# Plotting Primary and Secondary Growth
plt.figure(figsize=(10, 6))

# Plot Primary Growth (Total Weight Lifted)
plt.plot(df['Week'], df['Total_Weight'], label='Primary Growth (Total Weight Lifted)', color='Red', marker='o')

# Plot Secondary Growth (Estimated 1RM)
plt.plot(df['Week'], df['Estimated_1RM'], label='Secondary Growth (Estimated 1RM)', color='Green', marker='o')

# Adding Titles and Labels
plt.title('Combined Primary and Secondary Growth Over Weeks')
plt.xlabel('Week')
plt.ylabel('Weight (kg)')
plt.legend()
plt.grid(True)

# Display the plot
plt.show()





# Data for primary growth (total weight lifted)
data = {
    'Week': [1, 2, 3, 4, 5, 6],
    'Squad':[50,100,150,200,250,300],
    'Estimated_1RM':[60,100,150,200,250,300]
}
df = pd.DataFrame(data)
plt.figure(figsize=(10,6))

plt.plot(df['Week'], df['Squad'], label='squad', color='Red', marker='o')
plt.plot(df['Week'],df['Estimated_1RM'], label='Secondary Growth', color ='Green',marker ='o')
plt.title('Growth of 5 weeks')
plt.xlabel('Week')
plt.ylabel('Load(S)')
plt.legend()
plt.grid(True)




data = {
    'Week': [1, 2, 3, 4, 5, 6],
    'Bench':[50,100,150,200,250,300],
    'Estimated_1RM':[60,100,150,200,250,300]
}
df = pd.DataFrame(data)
plt.figure(figsize=(10,6))

plt.plot(df['Week'], df['Bench'], label='Bench', color='Blue', marker='o')
plt.plot(df['Week'],df['Estimated_1RM'], label='Secondary Growth', color ='Green',marker ='o')
plt.title('Growth of 5 weeks')
plt.xlabel('Week')
plt.ylabel('Load(S)')
plt.legend()
plt.grid(True)




data = {
    'Week': [1, 2, 3, 4, 5, 6],
    'Deadlift':[50,100,150,200,250,300],
     'Estimated_1RM':[60,100,150,200,250,300]
}
df = pd.DataFrame(data)
plt.figure(figsize=(10,6))

plt.plot(df['Week'], df['Deadlift'], label='Deadlift', color='Red', marker='o')
plt.plot(df['Week'],df['Estimated_1RM'], label='Secondary Growth', color ='Green',marker ='o')
plt.title('Growth of 5 weeks')
plt.xlabel('Week')
plt.ylabel('Load(S)')
plt.legend()
plt.grid(True)

plt.show()
```
