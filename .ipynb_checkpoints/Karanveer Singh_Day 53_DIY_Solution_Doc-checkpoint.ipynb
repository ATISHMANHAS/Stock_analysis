```python
import pandas as pd

df = pd.read_csv("Credit Card Customer Data.csv")
print(df.head())
```

```python
missing_values = df.isnull().sum()
print("Missing Values:\n", missing_values)

df = df.drop(['UnnecessaryColumn1', 'UnnecessaryColumn2'], axis=1, errors='ignore')

df = df.dropna()

print("\nData after handling missing values:")
print(df.info())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

df.hist(bins=20, figsize=(15, 10))
plt.suptitle("Univariate Analysis - Distribution of each variable")
plt.show()



```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
scaled_data = scaler.fit_transform(df)

scaled_df = pd.DataFrame(scaled_data, columns=df.columns)

print("Standardized Data:\n", scaled_df.head())
```

```python
from sklearn.cluster import KMeans

wcss = []
K = range(1, 11)

for k in K:
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(scaled_df)
    wcss.append(kmeans.inertia_)  

plt.figure(figsize=(8, 5))
plt.plot(K, wcss, 'bo-')
plt.xlabel("Number of Clusters (K)")
plt.ylabel("Within-Cluster Sum of Squares (WCSS)")
plt.title("Elbow Method for Optimal K")
plt.show()
```

```python
from sklearn.metrics import silhouette_score

silhouette_scores = []
for k in range(2, 11):
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(scaled_df)
    score = silhouette_score(scaled_df, kmeans.labels_)
    silhouette_scores.append(score)

plt.figure(figsize=(8, 5))
plt.plot(range(2, 11), silhouette_scores, 'go-')
plt.xlabel("Number of Clusters (K)")
plt.ylabel("Silhouette Score")
plt.title("Silhouette Score for Optimal K")
plt.show()
```

```python
optimal_k = 3  
final_model = KMeans(n_clusters=optimal_k, random_state=42)
final_model.fit(scaled_df)

df['Cluster'] = final_model.labels_

for column in df.columns[:-1]:  
    plt.figure(figsize=(8, 5))
    sns.boxplot(x='Cluster', y=column, data=df)
    plt.title(f"Cluster Behavior for {column}")
    plt.show()
```

```python

centroids = final_model.cluster_centers_

final_silhouette_score = silhouette_score(scaled_df, final_model.labels_)

print("Centroids of final clusters:\n", centroids)
print("\nSilhouette Score for the final model:", final_silhouette_score)
```
