```python
import pandas as pd

df = pd.read_csv("vehicle.csv")

df = df.dropna()
df
```

```python
target = df["class"]
features = df.drop("class", axis=1)
df
```

```python
from sklearn.preprocessing import StandardScaler


scaler = StandardScaler()

scaled_features = scaler.fit_transform(features)
```

```python
from sklearn.decomposition import PCA

pca = PCA(n_components=2)

pca_features = pca.fit_transform(scaled_features)
pca_features
```

```python
final_df = pd.DataFrame(data=pca_features, columns=["PC1", "PC2"])
final_df["class"] = target.values
final_df["class"]
```

```python
import matplotlib.pyplot as plt


plt.figure(figsize=(10, 6))
for c in final_df["class"].unique():
    subset = final_df[final_df["class"] == c]
    plt.scatter(subset["PC1"], subset["PC2"], label=c, alpha=0.7)

plt.xlabel("Principal Component 1")
plt.ylabel("Principal Component 2")
plt.legend(title="Class")
plt.title("PCA of Vehicle Dataset")
plt.show()
```

```python
variance_ratio = pca.explained_variance_ratio_

print("Variance ratio of the 2-D data using PCA:", variance_ratio)
print("Total variance explained:", variance_ratio.sum())
```
