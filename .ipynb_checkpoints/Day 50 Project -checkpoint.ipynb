```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import AgglomerativeClustering
from sklearn.metrics import mean_squared_error
from sklearn.decomposition import PCA
```

```python
df = pd.read_csv('zoo.csv')
print(df.info())
print(df.head())
```

```python
unique_classes = df.iloc[:, -1].unique()
num_unique_classes = len(unique_classes)
print("Unique Classes:", unique_classes)
print("Number of Classes:", num_unique_classes)
```

```python
X = df.iloc[:, 1:-1]  # exclude animal name and class
y_true = df.iloc[:, -1]  # actual class label
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
from sklearn.cluster import AgglomerativeClustering
model = AgglomerativeClustering(n_clusters=num_unique_classes)
y_pred = model.fit_predict(X_scaled)
```

```python
mse = mean_squared_error(y_true, y_pred)
print("Mean Squared Error:", mse)
```

```python
from sklearn.metrics import mean_squared_error
mse = mean_squared_error(y_true, y_pred)
print("Mean Squared Error:", mse)
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
plot_df = pd.DataFrame(X_pca, columns=['PCA1', 'PCA2'])
plot_df['Predicted_Cluster'] = y_pred
plot_df['True_Class'] = y_true

sns.set(style="whitegrid", palette="muted")
plt.figure(figsize=(12, 5))

# Plot Predicted Clusters
plt.subplot(1, 2, 1)
sns.scatterplot(data=plot_df, x='PCA1', y='PCA2', hue='Predicted_Cluster', palette='Set1')
plt.title('Agglomerative Clustering Result')
plt.legend(title='Cluster')

# Plot Actual Class Labels
plt.subplot(1, 2, 2)
sns.scatterplot(data=plot_df, x='PCA1', y='PCA2', hue='True_Class', palette='Set2')
plt.title('Actual Class Labels')
plt.legend(title='Class')

plt.tight_layout()
plt.show()
```
