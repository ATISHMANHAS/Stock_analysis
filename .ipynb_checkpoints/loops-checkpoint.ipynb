```python
class car:
    def __init__(self,name,color,mileage):
         self.color=color
         self.modelname = name
         self.mileage = mileage
    def display(self):
        print("I have ",self.modelname," car of",self.color," colour . It gives mileage of ",self.mileage)


x= car('BMW','Blue',80)

x.display()
```

```python
# Inheritence

class bus(car):
    def __init__(self,name,color,mileage,year):
         self.color=color
         self.modelname = name
         self.mileage = mileage
         self.year=year
    def model(self):
        print("I have ",self.modelname," car of",self.year,' ',self.color," colour . It gives mileage of ",self.mileage)

y= bus('BMW','Blue',70,2020)

y.display()
y.model()
```

```python
def SUM(a):
    if a==0:
        return 1
    else:
        return a+SUM(a-1)
```

```python
print(SUM(10))
SUM(20)
for i in range(1,11):
    print(SUM(i))
```

```python
def fact(c):
    if c==1:
        return 1
    else:
        return c*fact(c-1)
```

```python
fact(10)
for i in range(1,11):
    print(fact(i))
```

```python
def fab_sum(n):
    a=0
    b=1
    if n==1:
        return 0
    elif n==2:
        return 1
    else:
        return fab_sum(n-1)+fab_sum(n-2)
    

```

```python
for i in range(1,11):
    print(fab_sum(i))
```

```python
import pandas as pd
import numpy as np

# Load datasets
movies_df = pd.read_csv('movies.csv')
ratings_df = pd.read_csv('ratings.csv')

# Preview the datasets
print(movies_df.head())
print(ratings_df.head())

```

```python
# Check for missing values
print(movies_df.isnull().sum())
print(ratings_df.isnull().sum())

# Distribution of ratings
import matplotlib.pyplot as plt
import seaborn as sns

sns.histplot(ratings_df['rating'], bins=10, kde=True)
plt.title('Distribution of Ratings')
plt.xlabel('Rating')
plt.ylabel('Frequency')
plt.show()

# Unique users and movies
num_users = ratings_df['userId'].nunique()
num_movies = movies_df['movieId'].nunique()
print(f'Unique Users: {num_users}, Unique Movies: {num_movies}')

# Average rating and total movies at genre level
genre_counts = movies_df['genres'].str.get_dummies(sep='|').sum()
average_ratings = ratings_df.groupby('movieId')['rating'].mean()

# Combine average ratings with movies
movies_df['average_rating'] = movies_df['movieId'].map(average_ratings)
genre_average = movies_df.groupby('genres')['average_rating'].mean()

print(genre_average)
print(f'Unique Genres: {len(genre_counts)}')
```

```python
def popularity_based_recommender(genre, min_reviews, n):
    # Filter movies by genre
    genre_filtered = movies_df[movies_df['genres'].str.contains(genre)]
    
    # Join with ratings to count reviews
    movie_review_counts = ratings_df.groupby('movieId').size().reset_index(name='num_reviews')
    genre_filtered = genre_filtered.merge(movie_review_counts, on='movieId')
    
    # Filter based on minimum reviews
    popular_movies = genre_filtered[genre_filtered['num_reviews'] >= min_reviews]
    
    # Sort by average rating and get top N
    recommendations = popular_movies.sort_values(by='average_rating', ascending=False).head(n)
    return recommendations[['title', 'average_rating', 'num_reviews']]

# Example usage
print(popularity_based_recommender('Comedy', 100, 5))
```

```python
def content_based_recommender(movie_title, n):
    # Find the target movie
    target_movie = movies_df[movies_df['title'] == movie_title]
    
    if target_movie.empty:
        return "Movie not found."
    
    target_genres = target_movie.iloc[0]['genres']
    
    # Recommend movies with similar genres
    recommended_movies = movies_df[movies_df['genres'].str.contains(target_genres) & (movies_df['title'] != movie_title)]
    
    # Sort by average rating and get top N
    recommendations = recommended_movies.sort_values(by='average_rating', ascending=False).head(n)
    return recommendations[['title', 'average_rating']]

# Example usage
print(content_based_recommender('Toy Story', 5))

```

```python
from sklearn.metrics.pairwise import cosine_similarity

def collaborative_filtering_recommender(user_id, n, threshold):
    # Create a user-item matrix
    user_movie_ratings = ratings_df.pivot_table(index='userId', columns='movieId', values='rating').fillna(0)
    
    # Compute the cosine similarity matrix
    similarity_matrix = cosine_similarity(user_movie_ratings)
    
    # Create a DataFrame for similarity
    similarity_df = pd.DataFrame(similarity_matrix, index=user_movie_ratings.index, columns=user_movie_ratings.index)
    
    # Find similar users
    similar_users = similarity_df[user_id].sort_values(ascending=False).head(threshold).index
    
    # Get movies rated by similar users
    similar_users_ratings = user_movie_ratings.loc[similar_users]
    
    # Recommend movies that the target user hasn't rated yet
    recommendations = similar_users_ratings.loc[:, similar_users_ratings.loc[user_id] == 0].mean(axis=0).sort_values(ascending=False).head(n)
    
    # Map to movie titles
    recommended_movie_ids = recommendations.index
    recommended_movies = movies_df[movies_df['movieId'].isin(recommended_movie_ids)]
    return recommended_movies[['title', 'average_rating']]

# Example usage
print(collaborative_filtering_recommender(1, 5, 100))
```

```python
import ipywidgets as widgets
from IPython.display import display

# Popularity-based recommender GUI
genre_input = widgets.Text(description='Genre:')
min_reviews_input = widgets.IntText(description='Min Reviews:')
num_recommendations_input = widgets.IntText(description='Num Recommendations:')
button = widgets.Button(description='Get Recommendations')

def on_button_click(b):
    recommendations = popularity_based_recommender(genre_input.value, min_reviews_input.value, num_recommendations_input.value)
    display(recommendations)

button.on_click(on_button_click)
display(genre_input, min_reviews_input, num_recommendations_input, button)
```

```python
def mean(arr, ln):
    value = 0
    for i in arr:
        value += i
    mean_value = value / ln
    return mean_value

def variance(arr):
    arr2 = []
    l = len(arr)
    m = mean(arr, l)
    
    for x in arr:
        arr2.append((x - m) ** 2)
    
    # Calculate the variance
    variance_value = sum(arr2) / l  # For population variance
    # variance_value = sum(arr2) / (l - 1)  # For sample variance, uncomment if needed
    return variance_value

a = [5, 2, 1, 3, 4]
print("Variance:", variance(a))
```
