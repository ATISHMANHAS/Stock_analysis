```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from scipy import stats
```

```python
df = pd.read_csv('exam.csv')
df
```

```python
scores = ['math score', 'reading score', 'writing score']
```

```python
plt.figure(figsize=(15, 5))

for i, score in enumerate(scores):
    plt.subplot(1, 3, i + 1)
    sns.histplot(df[score], kde=True)
    plt.title(f'Distribution of {score}')
    plt.xlabel(score)
    plt.ylabel('Frequency')

plt.tight_layout()
plt.show()

```

```python
mean_std_results = {}
for score in scores:
    mean = df[score].mean()
    std = df[score].std()
    mean_std_results[score] = (mean, std)
    print(f'Mean of {score}: {mean:.2f}, Standard Deviation of {score}: {std:.2f}')
```

```python
skew_kurt_results = {}
for score in scores:
    skewness = stats.skew(df[score])
    kurtosis = stats.kurtosis(df[score])
    skew_kurt_results[score] = (skewness, kurtosis)
    print(f'Skewness of {score}: {skewness:.2f}, Kurtosis of {score}: {kurtosis:.2f}')
```

```python
z_scores = {}
for score in scores:
    z_scores[score] = stats.zscore(df[score])
    df[f'z_{score}'] = z_scores[score]
```

```python
plt.figure(figsize=(15, 5))

for i, score in enumerate(scores):
    plt.subplot(1, 3, i + 1)
    sns.histplot(df[f'z_{score}'], kde=True)
    plt.title(f'Z-Distribution of {score}')
    plt.xlabel(f'z_{score}')
    plt.ylabel('Frequency')

plt.tight_layout()
plt.show()
```
