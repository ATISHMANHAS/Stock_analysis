```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
 Problem Statement: Web Scrapping using BeautifulSoup
```

```python
a=requests.get("https://en.wikipedia.org/wiki/List_of_countries_and_dependencies_by_population")
```

```python
src = a.content
document = BeautifulSoup(src, 'html.parser')
document.prettify()
```

```python
table=document.find("table")
table
```

```python

table.find('tr') 
```

```python
row=table.find_all("tr")
del row[0]
```

```python
header=[a.get_text() for a in table.find_all("th")]
header
```

```python
l=[]
for i in row:
    c=[a.get_text() for a in i.find_all("td")]
    l.append(c)
```

```python
l
```

```python
b=pd.DataFrame(l,columns=header)
```
