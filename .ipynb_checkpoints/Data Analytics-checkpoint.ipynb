```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load CSV files into DataFrames
amazon_data = pd.read_csv('amazon.csv')
ebay_data = pd.read_csv('Ecommerca Data chairsonebay - dataset_ebay-items-scraper_2024-11-18_10-32-06-483 (2).csv.csv')

# Display basic information
print("Amazon Data Sample:")
print(amazon_data.head())

print("\nEbay Data Sample:")
print(ebay_data.head())
```

```python
def clean_data(df):
    df = df.drop_duplicates()
    df = df.ffill().bfill()  # Updated to avoid FutureWarning
    numeric_cols = df.select_dtypes(include='object').columns
    for col in numeric_cols:
        df[col] = pd.to_numeric(df[col], errors='coerce')
    return df

amazon_data = clean_data(amazon_data)
ebay_data = clean_data(ebay_data)

# Confirm data is clean
print("\nMissing Values After Cleaning (Amazon):")
print(amazon_data.isnull().sum())

print("\nMissing Values After Cleaning (Ebay):")
print(ebay_data.isnull().sum())
```

```python
# Descriptive Statistics
print("\nAmazon Data Statistics:")
print(amazon_data.describe())

print("\nEbay Data Statistics:")
print(ebay_data.describe())
```

```python
# Sales Distribution (Using 'discounted_price' for Amazon and 'price' for eBay)
plt.figure(figsize=(10, 5))
sns.histplot(amazon_data['discounted_price'], kde=True, color='blue')
plt.title('Amazon Discounted Price Distribution')
plt.xlabel('Discounted Price')
plt.ylabel('Frequency')
plt.show()

plt.figure(figsize=(10, 5))
sns.histplot(ebay_data['price'], kde=True, color='green')
plt.title('Ebay Price Distribution')
plt.xlabel('Price')
plt.ylabel('Frequency')
plt.show()
```

```python
# Correlation Heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(amazon_data.corr(), annot=True, cmap='coolwarm')
plt.title('Amazon Data Correlation Heatmap')
plt.show()

plt.figure(figsize=(10, 6))
sns.heatmap(ebay_data.corr(), annot=True, cmap='coolwarm')
plt.title('Ebay Data Correlation Heatmap')
plt.show()
```
