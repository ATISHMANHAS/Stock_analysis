```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv('winequality-red.csv')

df.head()
```

```python

filtered_df = df[df['quality'].isin([7, 8])]
filtered_df
```

```python

sample = filtered_df.sample(n=10, random_state=1)  # Set a seed for reproducibility
sample_alcohol = sample['alcohol']
print("Sample Alcohol Content:\n", sample_alcohol)
```

```python

mean_sample = sample_alcohol.mean()
std_sample = sample_alcohol.std()
print("Mean:", mean_sample)
print("Standard Deviation:", std_sample)
```

```python

sns.histplot(sample_alcohol, kde=True, stat='density', bins=10)
plt.title('Alcohol Content in Sample')
plt.xlabel('Alcohol Content')
plt.ylabel('Density')
plt.show()
```

```python

n = len(sample_alcohol)
standard_error = std_sample / np.sqrt(n)
print("Standard Error:", standard_error)
```

```python
Sampling Distribution Tasks
```

```python

means = []
stds = []
for _ in range(9):
    sample = filtered_df.sample(n=10, random_state=None)
    means.append(sample['alcohol'].mean())
    stds.append(sample['alcohol'].std())

print("Means of Samples:", means)
print("Standard Deviations of Samples:", stds)
```

```python

plt.figure(figsize=(10, 6))
sns.histplot(means, kde=True, stat='density', bins=9)
plt.title('Means of Sampling Distribution')
plt.xlabel('Mean Alcohol Content')
plt.ylabel('Density')
plt.show()
```

```python

std_sampling_distribution = np.std(means, ddof=1)
sampling_distribution_se = std_sampling_distribution / np.sqrt(len(means))
print("Standard Error of Sampling Distribution:", sampling_distribution_se)
```
