```python

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import DBSCAN

df = pd.read_csv("Mall_Customers.csv")

numeric_df = df.select_dtypes(include=[np.number])  # Select only numeric columns
correlation_matrix = numeric_df.corr()  # Calculate correlation
print("Correlation Matrix:\n", correlation_matrix)

least_correlated_col = correlation_matrix.mean().idxmin()
print(f"Dropping column with least correlation: {least_correlated_col}")
df = df.drop(columns=[least_correlated_col])

if 'Gender' in df.columns:
    df = pd.get_dummies(df, columns=['Gender'], drop_first=True)

X = df[['Annual Income (k$)', 'Spending Score (1-100)']].values  # Feature selection

dbscan = DBSCAN(eps=12.5, min_samples=4)
labels = dbscan.fit_predict(X)
df['Cluster'] = labels

cluster_sizes = df['Cluster'].value_counts()
print("Cluster Sizes:\n", cluster_sizes)

plt.scatter(
    df['Annual Income (k$)'],
    df['Spending Score (1-100)'],
    c=df['Cluster'],
    cmap='viridis',
    s=50,
    alpha=0.7
)
plt.title('Annual Income vs Spending Score with DBSCAN Clusters')
plt.xlabel('Annual Income (k$)')
plt.ylabel('Spending Score (1-100)')
plt.colorbar(label='Cluster')
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from fcmeans import FCM

df = pd.read_csv("Mall_Customers.csv")
X = df[['Age', 'Spending Score (1-100)']].to_numpy()

n_clusters = 2
fcm = FCM(n_clusters=n_clusters)
fcm.fit(X)

centers = fcm.centers
memberships = fcm.u
df['Cluster'] = fcm.predict(X)

plt.figure(figsize=(8, 6))
for i in range(n_clusters):
    cluster_data = X[df['Cluster'] == i]
    plt.scatter(cluster_data[:, 0], cluster_data[:, 1], label=f'Cluster {i}', alpha=0.6, s=50)

plt.scatter(centers[:, 0], centers[:, 1], c='red', marker='x', s=200, label='Centers')
plt.title('Age vs Spending Score (Fuzzy C-Means)')
plt.xlabel('Age')
plt.ylabel('Spending Score (1-100)')
plt.legend()
plt.show()

spending_predictions = fcm.predict(X)
df['Spending Capability'] = memberships.max(axis=1)
grouped = df.groupby('Cluster')['Spending Capability'].mean()
highest_spending_cluster = grouped.idxmax()
print(f"Cluster with highest spending capability: {highest_spending_cluster}")

plt.figure(figsize=(8, 6))
for i in range(n_clusters):
    cluster_data = X[df['Cluster'] == i]
    spending_strength = memberships[df['Cluster'] == i, i]
    plt.scatter(cluster_data[:, 0], cluster_data[:, 1], c=spending_strength, cmap='coolwarm', s=50, label=f'Cluster {i}', alpha=0.6)

plt.colorbar(label='Spending Capability')
plt.title('Age vs Spending Score (Cluster Spending Capability)')
plt.xlabel('Age')
plt.ylabel('Spending Score (1-100)')
plt.legend()
plt.show()
```
